```python
import os
from nipype.interfaces import afni as afni
import nipype.pipeline.engine as pe 
import nipype.interfaces.io as nio
import nipype.interfaces.utility as util
data_dir = os.path.abspath('../DATA')
print data_dir
print os.listdir(data_dir)
```

```python
#myfiles = [os.path.abspath('../DATA/run001+orig.BRIK'),
#          os.path.abspath('../DATA/run002+orig.BRIK')]

myfiles = [os.path.abspath('../DATA/sub001/run001.nii.gz'),
          os.path.abspath('../DATA/sub001/run002.nii.gz')] 
```

```python
automask = pe.MapNode(interface=afni.Automask(),
                   name="automask",
                     iterfield=['in_file'])
automask.inputs.in_file = myfiles
```

```python
automask.inputs.dilate = 1
automask.base_dir = data_dir
automask.inputs.out_file = 'mask'
automask.inputs.brain_file = 'masked_run'
#automask.inputs.in_file = os.path.abspath('../DATA/run001+orig.BRIK')
automask.base_dir = data_dir
automask.inputs.outputtype = 'NIFTI_GZ'
```
